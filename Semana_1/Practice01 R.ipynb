# Tablas de frecuencia y Gráficas para variables Nominales u ordinales
Creando la variable ordinal clase.peso: clasificación según peso. Recilación nacidos con pesos inferiores a 100 onzas son considerados livianos, aquellos con pesos entre 100 (inclusive) y 140 son aceptables y con pesos de 140 o más se consideran sanos

```r
babies$clase.peso <- ifelse(babies$bwt < 100, 'Liviano', ifelse(babies$bwt < 140, 'Aceptable', 'Sano'))
```

Si la variable es cualitativa es ordinal, se debe indicar el orden

```r
babies$clase.peso <- factor(babies$clase.peso, levels = c('Liviano', 'Aceptable', 'Sano'))
```

```r
(Frec.clase.peso <- table(babies$clase.peso))
```

Tabla de frecuencias relativas para clase.peso

```r
(Frec.Rel.clase.peso <- prop.table(Frec.clase.peso))
```

Tabla de frecuencias absolutas y relativas para clase.peso

```r
cbind(Frec.clase.peso, Frec.Rel.clase.peso)
```

Diagrama de barras para clase.peso

```r
barplot(Frec.Rel.clase.peso)
```

Diagrama de barras para clase.peso mejorado

```r
barplot(Frec.Rel.clase.peso, main = 'Clasificación según el peso de los bebés',
        ylab = 'Frecuencia relativa', col = 'blue', las = 1, ylim = c(0, 0.8))
```

Diagrama de torta para clase.peso

```r
pie(Frec.Rel.clase.peso)
```

Diagrama de torta mejorado para clase.peso

```r
porcent = paste(round(Frec.Rel.clase.peso*100, 2), "%", sep = "")
newlabel = paste(row.names(Frec.Rel.clase.peso), "\n(", porcent, ")", sep = "")
pie(Frec.Rel.clase.peso, newlabel, radius = 1, main = 'ClasificaciÃ³n según el peso de los bebés')
```
